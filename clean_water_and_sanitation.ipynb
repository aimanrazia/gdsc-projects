```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Q = 1000  # flow rate (m³/h)
C0 = 500  # initial contaminant concentration (mg/m³)
k = 0.1  # rate constant (1/h)
t_final = 10  # final time (h)

# Time vector
t = np.linspace(0, t_final, 1000)

# Initial condition
C = C0 * np.ones_like(t)

# Solve the ODE using the Euler method
dt = t[1] - t[0]
for i in range(1, len(t)):
    dC_dt = -k * C[i-1]
    C[i] = C[i-1] + dC_dt * dt

# Plot the results
plt.plot(t, C)
plt.xlabel('Time (h)')
plt.ylabel('Contaminant concentration (mg/m³)')
plt.title('Water treatment simulation')
plt.show()
```
